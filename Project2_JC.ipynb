```python
import random #import random to randomize the word to guess

#function to create the random word you will be guessing
def get_word():
    words_list = ['Variable', 'List', 'Tuple', 'Dictionary', 'Function', 'Class', 'Inheritance',
                  'Module', 'Package', 'Loop', 'Conditional', 'Exception', 'Lambda', 'Decorator',
                  'Generator', 'Slice', 'Recursion', 'API', 'Framework', 'IDE']
    word = random.choice(words_list)
    return word.lower()

#this function is just a list of the displayed images that correlate to the game
def display(attempts):
    stages = [""" 
                --------
                |      |
                |      o
                |     \|/
                |      |
                |     / \\
                _
                """,
                """  
                --------
                |      |
                |      o
                |     \|/
                |      |
                |     /
                _
                """,
                """  
                --------
                |      |
                |      o
                |     \|/
                |      |
                |     
                _
                """,
                """  
                --------
                |      |
                |      o
                |      |/
                |      |
                |     
                _
                """,
                """  
                --------
                |      |
                |      o
                |      |
                |      |
                |     
                _
                """,
                """  
                --------
                |      |
                |      o
                |      |
                |      
                |     
                _
                """,
                """  
                --------
                |      |
                |      
                |      
                |      
                |     
                _
                """
    ]
    return stages[attempts]

#actual game code function
def play():
    #these are welcoming statements and category for the player
    print("Welcome to Hangman!")
    print("The category is Python")
    print("You have 6 attempts to guess the word.")
    #set of lists to be called on during the game
    guessed_letters = []
    guessed_words = []
    word_to_guess = get_word()
    attempts = 6
    guessed_word = ["_ "] * len(word_to_guess)

    while True:
        #prints the guessed letters and words and images
        print("Guessed letters: ", ", ".join(guessed_letters))
        print("Guessed words: ", ", ".join(guessed_words))
        print("".join(guessed_word))
        print(display(attempts))

        guess = input("Guess a letter or word: ").lower()
        #checking if guess letters are correct and appending them to the correlating lists
        if guess in guessed_letters:
            print("You already guessed that letter. Try again.")
            continue

        if len(guess) == 1 and guess.isalpha():
            if guess in word_to_guess:
                for index, letter in enumerate(word_to_guess):
                    if letter == guess:
                        guessed_word[index] = guess
            else:
                attempts -= 1
                print("Incorrect guess! Attempts remaining:", attempts)
        elif len(guess) == len(word_to_guess) and guess.isalpha():      #checking guessed words and apppending to list
            if guess in guessed_words:
                print("You've already guessed that word.")
            elif guess != word_to_guess:
                print(f"{guess} is not the word")
                attempts -= 1
                guessed_words.append(guess)
            else:
                print("Congratulations! You guessed the word!")
                break

        guessed_letters.append(guess)
        # ending the game due to no more guesses or getting the word
        if attempts == 0:
            print("Sorry, you ran out of attempts. The word was:", word_to_guess)
            print(display(attempts))
            break
        elif "_" not in "".join(guessed_word):
            print("Congratulations! You guessed the word!")
            break

if __name__ == "__main__":
    play()
```
